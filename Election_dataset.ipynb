```python
file = "mega_merged_df.csv"
df = pd.read_csv(file)
df = df.dropna(subset=['state_name'])
df.head()
```

```python
df_dummy = df.copy()
```

```python

df['state_name'] = df['state_name'].apply(lambda x: x.strip())
df['e_results'] = ''

for i in df.index:
    state = df.loc[i, 'state_name']
    try:
        df['e_results'].at[i] = elec_df1[elec_df1['state_name'].str.strip() == state.strip()]['elec_result'].values[0]
    except Exception as e:
        print(e)
    
df.sample(n=20)
```

```python
# Turn dataframe into a dictionary
election_series = elec_df[["state_name", "red_blue_2020"]]
key_value = election_series.set_index('state_name')['red_blue_2020'].to_dict()
print(key_value)
```

```python

d = key_value.items()

new_d = {str(key): str(value) for key, value in d}
print(new_d)
```

```python
df['elec_result'] = df['state_name'].map(new_d)
```
