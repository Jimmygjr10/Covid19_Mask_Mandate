```python
# read in the county_pop_merge.csv file

file = "Resources/county_pop_loc_sqlmerge1.csv"
df = pd.read_csv(file)
```

```python
df = df.rename({"state_name_x" : "state_name"}, axis = 1)
```

```python
df = df.dropna(subset=['state_name'])
df.head()
```

```python
# copy df to df_dummy

df_dummy = df.copy()
```

```python
# add in the election results to the mandate data

df['state_name'] = df['state_name'].apply(lambda x: x.strip())
df['e_results'] = ''

for i in df.index:
    state = df.loc[i, 'state_name']
    try:
        df['e_results'].at[i] = elec_df1[elec_df1['state_name'].str.strip() == state.strip()]['elec_result'].values[0]
    except Exception as e:
        print(e)
    
df.sample(n=20)
```

```python
#count the election results

df["e_results"].value_counts()
```

```python
# count the number of "state_name" and "e_result" values

df[["state_name", "e_results"]].value_counts()
```

```python
# calcualting the percentage of cases to the overall population

df['cases_perc'] = round(df['cases']/df['popestimate2020']*100)

df.head()
```

```python
df['cases_perc'].isna().sum()
```

```python
df.dropna(subset=['cases_perc'], inplace=True)
df.head()
```

```python
# turn the cases_perc column from float to int64

df = df.astype({'cases_perc': 'int64'})
df.dtypes
```
